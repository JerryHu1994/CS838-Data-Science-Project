# 1. Label the sample pairs

```python
import py_entitymatching as em
import pandas as pd
```

```python
S = em.read_csv_metadata("pairs_sample.csv", key='_id')
```

```python
S.head()
```

```python
G = S.copy(deep=True)
```

```python
G['label'] = G['ltable_vin'] == G['rtable_vin']
```

```python
G.head()
```

```python
G['label'].sum()
```

We label the car_pair to 1 for True, 0 for False.

```python
G['label'] = G['label'].apply(lambda x: 1 if x else 0)
```

```python
G.head()
```

```python
G.to_csv('pairs_sample_label.csv', index=False)
```
