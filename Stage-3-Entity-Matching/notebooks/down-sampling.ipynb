# 1. Down sampling

```python
import py_entitymatching as em
```

```python
cars_csv_file = "cars_com_clean.csv"
market_csv_file = "market_api_clean.csv"
```

```python
A = em.read_csv_metadata(cars_csv_file)
B = em.read_csv_metadata(market_csv_file)
```

```python
len(A), len(B)
```

```python
em.set_key(A, 'vin')
em.set_key(B, 'vin')
```

```python
sample_A, sample_B = em.down_sample(A, B, size=1500, y_param=2, show_progress=True)
```

```python
len(sample_A), len(sample_B)
```

```python
em.show_properties(sample_A)
```

# 2. Analyze sample results

```python
import pandas as pd
```

```python
sample_merge = pd.merge(sample_A, sample_B, how="inner", on=["vin"])
```
